```python
from numpy import array, linspace
from math import * #exp, log, sqrt, sin
from pylab import plot, xlabel, ylabel, show
```

```python
x_0 = 0.5
x = []
x.append(x_0)

for i in range(100):
    x_0 = 2-exp(-x_0)
    x.append(x_0)

plot(x)
show()
```

Como se observa en la gráfica anterior el valor de la función converge a 1.8 aproximadamente.

```python
x_0 = 0.5
x = []
x.append(x_0)

for i in range(100):
    x_0= exp(1- x_0**2)
    x.append(x_0)

plot(x)
show()
```

Para el caso de esta función no tenemos un número fijo, es decir, la función en este caso siempre diverge.

```python
x_0 = 0.5
x = []
x.append(x_0)

for i in range(100):
    x_0 = sqrt(1-log(x_0))
    x.append(x_0)

plot(x)
show()
```

```python

x_0 = 0
x_0 = 0.5
x = []
x.append(x_0)

for i in range(100):
    x_0= (x_0**2)+sin(2*x_0)
    x.append(x_0)

plot(x)
show()
```

Aquí cabe destacar que la función solo converge para 0, para valores distintos la función se encuentra fuera de rango

```python
x_0 = 0.5
x = []
x.append(x_0)

for i in range(100):
    x_0 = 0.5*asin(x_0-x_0**2)
    x.append(x_0)

plot(x)
show()
```
